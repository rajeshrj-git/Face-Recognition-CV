```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
# Load the model
model_loaded = load_model(r'C:\Users\rajes\Datascience_jp\Day28-Recognition & Transformers\Face_Recognition_CNN.h5')

# Dummy class indices; update this with actual indices used during training
class_indices = { 'Rajesh' :0,'Steve Rogers' : 1,'Tony Stark': 2}
class_indices = {v: k for k, v in class_indices.items()}
```

```python
def predict_face(image):
    image = cv2.resize(image, (150, 150))
    image = np.expand_dims(image, axis=0) / 255.0
    prediction = model_loaded.predict(image)
    class_index = np.argmax(prediction, axis=1)[0]
    class_label = class_indices.get(class_index, 'Unknown')
    return class_label
```

```python
# Load pre-trained face detector from OpenCV
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Open the webcam
video_capture = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()

    # Convert to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Loop over the detected faces
    for (x, y, w, h) in faces:
        # Draw a rectangle around the face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # Extract the face from the frame
        face = frame[y:y+h, x:x+w]
        face_label = predict_face(face)

        # Draw a label with the prediction
        cv2.rectangle(frame, (x, y-35), (x+w, y), (0, 255, 0), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, face_label, (x + 6, y - 6), font, 0.9, (255, 255, 255), 1)

    # Display the resulting frame
    cv2.imshow('Video', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
video_capture.release()
cv2.destroyAllWindows()
```
